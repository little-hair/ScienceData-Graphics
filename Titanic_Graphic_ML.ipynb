```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
sb.get_dataset_names()
```

```python
titanic = sb.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sb.pairplot(titanic, hue='class')
```

```python
# GRAPHIC AGE
age = titanic['age'].value_counts().sort_index()

axis_x = age.index
axis_y = age.values

plt.plot(axis_x, axis_y)

plt.xlabel('Ages')
plt.ylabel('amount')

plt.show()
```

```python
# GRAPHIC SEX
sex = titanic['sex'].value_counts()

axis_x = sex.index
axis_y = sex.values

bar_labels = ['Men', 'Woman']
bar_colors = ['blue', 'pink']

plt.bar(axis_x, axis_y, label=bar_labels, color=bar_colors)

plt.xlabel('Sex')
plt.ylabel('amount')

plt.title('GENDERS IN TITANIC')
plt.legend(title='Sex')

plt.show()
```

```python
# GRAPHIC MOST AGE amount IN TITATIC
age = titanic['age'].value_counts().nlargest(5)

axis_x = age.index
axis_y = age.values

bar_labels = ['1st', '2st', '3st', '4st', '5st']
bar_colors = ['magenta', 'green', 'purple', 'blue', 'pink']

plt.barh(axis_x, axis_y, label=bar_labels, color=bar_colors)
plt.title('AGES EMBARK')
plt.legend(title='Ages')

plt.show()
```

```python
# GRAPHIC EMBARK LOCAL

embark = titanic['embark_town'].value_counts()

axis_x = embark.index
axis_y = embark.values

bar_labels = ['Southampton', 'Cherboug', 'Queenstown']
bar_colors = ['red', 'orange', 'yellow']

plt.bar(axis_x, axis_y, label=bar_labels, color=bar_colors)

plt.title('EMBARK LOCAL')
plt.legend(title='Fruit color')

plt.ylabel('Amount People')

plt.show()
```

```python
# GRAPHIC ALIVE PEOLPLE

alive = titanic['alive'].value_counts().sort_index()

axis_x = alive.index
axis_y = alive.values

bar_labels = ['Dead', 'Alive']
bar_colors = ['red', 'green']

plt.bar(axis_x, axis_y, label=bar_labels, color=bar_colors)

plt.title('NUMBER OF SURVIVORS')
plt.legend(title='Survived')

plt.show()
```
